# Q1. Explain why we have to use the Exception class while creating a Custom Exception

When creating a custom exception in Python, it is recommended to inherit from the base Exception class or one of its derived classes. Here's an explanation of why we use the Exception class as the base for custom exceptions:

1.Consistency and Compatibility: The Exception class is the base class for all built-in exceptions in Python. By inheriting from the Exception class, you ensure that your custom exception follows the same inheritance hierarchy and behaves consistently with other exceptions in the language. It ensures compatibility with existing exception handling mechanisms in Python.

2.Standard Error Handling: In Python, exceptions are typically caught and handled using try-except blocks. By inheriting from Exception, your custom exception can be caught in a generic except block that handles all types of exceptions derived from Exception. This makes it easier to handle and manage exceptions in a standardized way.

3.Exception Hierarchy: The Exception class provides a hierarchy of built-in exception classes, each representing different types of errors. By using the Exception class as the base for your custom exception, you can position it within the existing exception hierarchy. This allows you to structure and organize your exceptions based on their relationships and provides a clear classification of error types.

4.Exception Handling Polymorphism: Inheriting from Exception allows your custom exception to benefit from polymorphic behavior. Polymorphism enables different exception types to be caught and handled in a uniform manner. You can catch your custom exception using a generic except block for Exception and handle it alongside other types of exceptions.

5.Future Compatibility: Python continues to evolve, and new features and improvements may be introduced in the language. By adhering to the established conventions, such as inheriting from Exception, your custom exception code remains compatible with future versions of Python and maintains a consistent and familiar structure for other developers.

# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans : LookupError This is the base class for those exceptions that are raised when a key or index used on a mapping or sequence is invalid or not found. The exceptions raised are :
 KeyError,
 IndexError

```python
# Index Error:
try:
    l=[3,4,5,6,7]
    print(l[5])
except LookupError as le:
    print(IndexError)
```

```python
# Key Error
try:
    dic={'manju':20,'sonam':12,'Divya':14}
    print(dic['Anshu'])
except LookupError as le:
    print(KeyError)
    
```
